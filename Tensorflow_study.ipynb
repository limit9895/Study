```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#스칼라 정의
a = tf.constant(1)
b = tf.constant(2)
print(a)
print(b)
print(tf.rank(a)) #차원 확인
a = tf.cast(a,tf.float32)
b = tf.cast(b,tf.float32)
print(a)
print(b)
#계산 : add(덧셈), subtract(뺄셈), multiply(곱셈), divide(나눗셈), mod(나눗셈_나머지), floordiv(나눗셈_몫),square(거듭제곱),sqrt(제곱근),
c = tf.math.add(a,b)
d = tf.math.square(c)
print(c)
print(tf.rank(c)) #차원 확인
print(d)
print(tf.rank(d)) #차원 확인
```

```python
#1차원 배열 정의
py_list = [10., 20., 30.] #파이썬 리스트
num_arr = np.array([10.,10.,10.]) #넘파이 배열

#텐서 변환(벡터)
vec1 = tf.constant(py_list, dtype = tf.float32)
vec2 = tf.constant(num_arr, dtype = tf.float32)

print(vec1)
print(vec2)
print(tf.rank(vec1))
print(tf.rank(vec2))

add1 = tf.math.add(vec1,vec2) #다른 연산 함수도 전부 사용 가능
print(add1)
print(tf.rank(add1))

add2 = vec1 + vec2 #다른 연산자도 전부 사용 가능
print(add2)
print(tf.rank(add2))

print(tf.reduce_sum(vec1)) #벡터를 구성하는 모든 원소들의 합 계산
print(tf.reduce_sum(vec2)) #벡터를 구성하는 모든 원소들의 합 계산

print(vec1 + 1) #브로드캐스팅 연산
```

```python
#2차원 배열 정의
list_of_list = [[10,20],[30,40]]

#텐서 변환(메트릭스)
mat1 = tf.constant(list_of_list)

print(mat1)
print(tf.rank(mat1))

#벡터 정의
vec1 = tf.constant([1, 0])
vec2 = tf.constant([-1, 2])

#텐서 변환(메트릭스)_stack
mat2 = tf.stack([vec1,vec2])

print(mat2)
print(tf.rank(mat2))

#연산
element_mul = tf.math.multiply(mat1,mat2)
print(element_mul)
print(tf.rank(element_mul))

#연산_브로드 캐스팅
element_bc = tf.math.multiply(mat1,3)
print(element_bc)
print(tf.rank(element_bc))

#행렬곱 연산(선형대수)
mat_mul = tf.matmul(mat1,mat2)
print(mat_mul)
print(tf.rank(mat_mul))

#넘파이 배열로 변환
np_arr = mat_mul.numpy()
print(type(np_arr))
print(np_arr)
```

```python
#자동미분
g = tf.random.Generator.from_seed(2020)
x = g.normal(shape = (10,))
y = 3*x-2

print('x :',x.numpy())
print('y :',y.numpy())

#손실 함수 정의
def cal_mse(x,y,a,b):
    y_pred = a*x+b
    squared_error = (y_pred - y)**2
    mean_squared_error = tf.reduce_mean(squared_error)
    return mean_squared_error


a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHS = 200

for epoch in range(1, EPOCHS + 1):
    with tf.GradientTape() as tape:
        mse = cal_mse(x,y,a,b)
    grad = tape.gradient(mse,{'a':a,'b':b})
    d_a,d_b = grad['a'],grad['b']
    
    a.assign_sub(d_a*0.05)
    b.assign_sub(d_b*0.05)
    
    if epoch%20 == 0:
        print('EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
    



```

```python
#자동미분
g = tf.random.Generator.from_seed(2020)
x = g.normal(shape = (10,))
y = 3*x-2

print('x :',x.numpy())
print('y :',y.numpy())

#손실 함수 정의
def cal_mse(x,y,a,b):
    y_pred = a*x+b
    squared_error = (y_pred - y)**2
    mean_squared_error = tf.reduce_mean(squared_error)
    return mean_squared_error


a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHS = 200

for epoch in range(1, EPOCHS + 1):
    with tf.GradientTape() as tape:
        mse = cal_mse(x,y,a,b)
    grad = tape.gradient(mse,{'a':a,'b':b})
    d_a,d_b = grad['a'],grad['b']
    print('1    EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
    
    a.assign_sub(d_a*0.05)
    b.assign_sub(d_b*0.05)

    print('2    EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
```

```python
#경사하강법 실습

#샘플에 활용할 데이터 셋 만들기
def make_linear(w, b, size,noise):
    x = np.random.rand(size)
    y = w*x+b
    noise = np.random.uniform(-abs(noise),abs(noise),size = y.shape)
    yy = y + noise
    plt.figure(figsize=(10, 7))
    plt.plot(x,y,color='r',label=f'y={w}*x+{b}')
    plt.scatter(x,yy,label ='data')
    plt.legend(fontsize=20)
    plt.show()
    print(f'w:{w},b:{b}')
    return x,yy

x,y = make_linear(w=0.3,b=0.5,size=100,noise=0.01)

#최대 반복 횟수
num_epoch = 100

#학습률(learnig_rate)
learning_rate = 0.005

#에러기록
errors = []

#random 한 값으로 w,b를 초기화
w = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low = 0.0, high = 1.0)

for epoch in range(num_epoch):
    #Hypothesis 정의
    y_hat = w*x+b
    
    #Loss Function 정의
    error = 0.5*((y_hat - y)**2).sum()
    if error < 0.005:
        break
    #Gradient 미분 계산
    w = w - learning_rate*((y_hat - y)*x).sum()
    b = b - learning_rate*(y_hat -y).sum()
    
    errors.append(error)
    
    if epoch%5 == 0:
        print("{0:2}     w = {1:.5f}, b = {2:.5f}     error = {3:.5f}".format(epoch, w, b, error))

print("----" * 15)
print("{0:2}     w = {1:.1f}, b = {2:.1f}     error = {3:.5f}".format(epoch, w, b, error))

plt.figure(figsize = (10,7))
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

#케라스의 내장 데이터셋에서 mnist 다운
mnist = tf.keras.datasets.mnist

#load_data()로 데이터셋 로드
(x_train, y_train),(x_test,y_test) = mnist.load_data()

#로드 된 데이터셋 확인
print('train set', x_train.shape, y_train.shape)
print('test set', x_test.shape, y_test.shape)

#데이터 시각화
#canvas 생성
fig, axes = plt.subplots(3,5)
fig.set_size_inches(8,5)

for i in range(15):
    ax = axes[i//5,i%5]
    #imshow 로 이미지 시각화
    ax.imshow(x_train[i], cmap='gray')
    ax.axis('off')
    ax.set_title(str(y_train[i]))

plt.tight_layout()
plt.show()
```

```python
#데이터 전처리(정규화)
#데이터 전처리 전 데이터의 형태 확인
x_train[0, 10:15, 10:15,]

print(f'정규화 전] 최소값:{x_train.min()}, 최대값:{x_train.max()}')

#데이터 정규화(0~1사이의 값으로 정규화)
x_train = x_train/x_train.max()
print(f'정규화 후] 최소값:{x_train.min()}, 최대값:{x_train.max()}')

#테스트 데이터 셋에도 동일하게 정규화
x_test = x_test/x_test.max()

#변환 후 x_train 배열의 데이터 확인
x_train[0, 10:15, 10:15,]

#데이터 정규화를 하는 이유는 입력데이터가 정규화 되어 모델이 학습하는 경우에 경사하강법 알고리즘에 의한 수렴속도가 빠르고, 국소 최적(local optimum)에 빠지는 현상을 방지해주기 때문이다.
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

mnist = tf.keras.datasets.mnist # mnist 로드
(x_train, y_train),(x_test,y_test) = mnist.load_data() #load_data()로 데이터셋 로드
x_train = x_train/x_train.max()#데이터셋 정규화
x_test = x_test/x_test.max()#데이터셋 정규화

#배치 정규화 및 활성화 함수
#각 레이어에서 활성화 함수 적용 전 미니 배치의 스케일을 정규화 → 데이터 전달되기전에 스케일 조정으로 안정적인 훈련이 가능하며 성능을 크게 향상시킴
#모델A : Dense + ReLU
model_a = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(32,activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
])
#print('ReLU')
#print(model_a.summary(),end = '\n\n\n')

#모델B : Dense + BatchNorm + ReLU
model_b = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(64),
        tf.keras.layers.BatchNormalization(),#배치 정규화
        tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
        tf.keras.layers.Dense(32),
        tf.keras.layers.BatchNormalization(),#배치 정규화
        tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
        tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU')
#print(model_b.summary(),end = '\n\n\n')

#모델C : LeakyReLU(0.2) + BatchNorm + Dense
model_c = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.LeakyReLU(alpha=0.2),#배치 정규화 후 활성화 함수로 LeakyReLU 적용
          tf.keras.layers.Dense(32),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.LeakyReLU(alpha=0.2),#배치 정규화 후 활성화 함수로 LeakyReLU 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + LeakyReLU')
#print(model_c.summary())

#모델D : ReLU + BatchNorm + Dense + 초기화방법 변경(GlortUniform → HeNormal) + L1규제 + 드롭아웃
he_normal = tf.keras.initializers.HeNormal
regularizer = tf.keras.regularizers.l1(l1=0.1)
model_d = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64,kernel_initializer=he_normal,kernel_regularizer=regularizer),
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32,kernel_initializer=he_normal,kernel_regularizer=regularizer),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + GlortUniform → HeNormal + L1규제 + 드롭아웃')
#print(model_d.summary())

#모델E : ReLU + BatchNorm + Dense + L1규제 + 드롭아웃
regularizer = tf.keras.regularizers.l1(l1=0.1)
model_e = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64,kernel_regularizer=regularizer),
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32,kernel_regularizer=regularizer),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + L1규제 + 드롭아웃')
#print(model_e.summary())

#모델F : ReLU + BatchNorm + Dense + L2규제 + 드롭아웃
regularizer = tf.keras.regularizers.l2(l2=0.1)
model_f = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64,kernel_regularizer=regularizer),
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32,kernel_regularizer=regularizer),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + L2규제 + 드롭아웃')
#print(model_f.summary())

#모델G : ReLU + BatchNorm + Dense + 초기화방법 변경(GlortUniform → HeNormal) + 드롭아웃
he_normal = tf.keras.initializers.HeNormal
model_g = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64,kernel_initializer=he_normal),
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32,kernel_initializer=he_normal),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + GlortUniform → HeNormal + 드롭아웃')
#print(model_g.summary())

#모델H : ReLU + BatchNorm + Dense + 드롭아웃
model_h = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64),
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Dropout(0.25), #25%의 노드를 임의로 삭제
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + dropout')
#print(model_h.summary())

#모델I : ReLU + BatchNorm + Dense + 초기화방법 변경(GlortUniform → HeNormal)
he_normal = tf.keras.initializers.HeNormal
model_i = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
          tf.keras.layers.Dense(64,kernel_initializer=he_normal),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(32,kernel_initializer=he_normal),
          tf.keras.layers.BatchNormalization(),#배치 정규화
          tf.keras.layers.Activation('relu'),#배치 정규화 후 활성화 함수 적용
          tf.keras.layers.Dense(10, activation='softmax')
])

#print('BatchNorm + ReLU + GlortUniform → HeNormal')
#print(model_i.summary())

#컴파일
model_a.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_b.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_c.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_d.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_e.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_f.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_g.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_h.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model_i.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])

#훈련
history_a = model_a.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_b = model_b.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_c = model_c.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_d = model_d.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_e = model_e.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_f = model_f.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_g = model_g.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_h = model_h.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
history_i = model_i.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=20)
```

```python
#콜백
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

mnist = tf.keras.datasets.mnist # mnist 로드
(x_train, y_train),(x_test,y_test) = mnist.load_data() #load_data()로 데이터셋 로드
x_train = x_train/x_train.max()#데이터셋 정규화
x_test = x_test/x_test.max()#데이터셋 정규화

model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(256,activation='relu'),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(32,activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
        ])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
```

```python
#체크포인트 설정 : 학습중에 가장 낮은 타겟값을 저장하는 방식
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath='tmp_checkpoint.ckpt', #체크포인트의 저장 경로 지정
                                                save_weights_only = True, #가중치만 저장할지 여부 설정
                                                save_best_only=True, #monitor 기준으로 가장 높은 epoch만 저장할지 모든 epoch를 저장 할 지 선택
                                                monitor = 'val_loss', #저장 시 기준이 되는 지표 설정 : val_loss지정시 검증 손실이 가장 낮은 epoch의 가중치 저장
                                                verbose = 1) #1로 설정시 매 epoch별 저장 여부를 알려주는로그 메시지 출력
model.fit(x_train,y_train,
          validation_data=(x_test,y_test),
          epochs=10,
          callbacks=[checkpoint]
          )
```

```python
#모델 체크포인트 로드 전
loss, acc = model.evaluate(x_test,y_test)
print(f'체크포인트 로드전 - loss : {loss:3f}, acc : {acc:3f}')

#모델 체크포인트 로드 후
model.load_weights('tmp_checkpoint.ckpt')
loss, acc = model.evaluate(x_test,y_test)
print(f'체크포인트 로드후 - loss : {loss:3f}, acc : {acc:3f}')
```

```python
#Early Stopping : 학습을 해도 원하는 타겟 값이 하강하지 않는다면 학습을 조기종료 시킴
model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(256,activation='relu'),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(32,activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
        ])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])

#Early Stopping 콜백 생성
earlystopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience = 3) #3 epoch 이내에 val_loss 값이 하강하지 않는다면 훈련 종료
model.fit(x_train,y_train,
          validation_data=(x_test,y_test),
          epochs=20,
          callbacks=[earlystopping]
          )
```

```python
#학습률 스케줄러(Learnig Rate Scheduler)
def scheduler(epoch,lr):
    tf.print(f'learning_rate : {lr:.5f}')
    #첫 5 epoch 동안 학습률 유지
    if epoch < 5:
        return lr
    else:
    #학습률 감소 적용
        return lr * tf.math.exp(-0.1)

#콜백 객체 생성 및 scheduler 함수 적용
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(scheduler)

model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(256,activation='relu'),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(32,activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
        ])

model.compile(tf.keras.optimizers.SGD(),
              loss='sparse_categorical_crossentropy',metrics =['accuracy'])

print(round(model.optimizer.lr.numpy(),5))

model.fit(x_train,y_train,
          validation_data=(x_test,y_test),
          epochs=10,
          callbacks=[lr_scheduler]
          )
print(round(model.optimizer.lr.numpy(),5))
```
