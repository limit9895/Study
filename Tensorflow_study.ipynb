```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#스칼라 정의
a = tf.constant(1)
b = tf.constant(2)
print(a)
print(b)
print(tf.rank(a)) #차원 확인
a = tf.cast(a,tf.float32)
b = tf.cast(b,tf.float32)
print(a)
print(b)
#계산 : add(덧셈), subtract(뺄셈), multiply(곱셈), divide(나눗셈), mod(나눗셈_나머지), floordiv(나눗셈_몫),square(거듭제곱),sqrt(제곱근),
c = tf.math.add(a,b)
d = tf.math.square(c)
print(c)
print(tf.rank(c)) #차원 확인
print(d)
print(tf.rank(d)) #차원 확인
```

```python
#1차원 배열 정의
py_list = [10., 20., 30.] #파이썬 리스트
num_arr = np.array([10.,10.,10.]) #넘파이 배열

#텐서 변환(벡터)
vec1 = tf.constant(py_list, dtype = tf.float32)
vec2 = tf.constant(num_arr, dtype = tf.float32)

print(vec1)
print(vec2)
print(tf.rank(vec1))
print(tf.rank(vec2))

add1 = tf.math.add(vec1,vec2) #다른 연산 함수도 전부 사용 가능
print(add1)
print(tf.rank(add1))

add2 = vec1 + vec2 #다른 연산자도 전부 사용 가능
print(add2)
print(tf.rank(add2))

print(tf.reduce_sum(vec1)) #벡터를 구성하는 모든 원소들의 합 계산
print(tf.reduce_sum(vec2)) #벡터를 구성하는 모든 원소들의 합 계산

print(vec1 + 1) #브로드캐스팅 연산
```

```python
#2차원 배열 정의
list_of_list = [[10,20],[30,40]]

#텐서 변환(메트릭스)
mat1 = tf.constant(list_of_list)

print(mat1)
print(tf.rank(mat1))

#벡터 정의
vec1 = tf.constant([1, 0])
vec2 = tf.constant([-1, 2])

#텐서 변환(메트릭스)_stack
mat2 = tf.stack([vec1,vec2])

print(mat2)
print(tf.rank(mat2))

#연산
element_mul = tf.math.multiply(mat1,mat2)
print(element_mul)
print(tf.rank(element_mul))

#연산_브로드 캐스팅
element_bc = tf.math.multiply(mat1,3)
print(element_bc)
print(tf.rank(element_bc))

#행렬곱 연산(선형대수)
mat_mul = tf.matmul(mat1,mat2)
print(mat_mul)
print(tf.rank(mat_mul))

#넘파이 배열로 변환
np_arr = mat_mul.numpy()
print(type(np_arr))
print(np_arr)
```

```python
#자동미분
g = tf.random.Generator.from_seed(2020)
x = g.normal(shape = (10,))
y = 3*x-2

print('x :',x.numpy())
print('y :',y.numpy())

#손실 함수 정의
def cal_mse(x,y,a,b):
    y_pred = a*x+b
    squared_error = (y_pred - y)**2
    mean_squared_error = tf.reduce_mean(squared_error)
    return mean_squared_error


a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHS = 200

for epoch in range(1, EPOCHS + 1):
    with tf.GradientTape() as tape:
        mse = cal_mse(x,y,a,b)
    grad = tape.gradient(mse,{'a':a,'b':b})
    d_a,d_b = grad['a'],grad['b']
    
    a.assign_sub(d_a*0.05)
    b.assign_sub(d_b*0.05)
    
    if epoch%20 == 0:
        print('EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
    



```

```python
#자동미분
g = tf.random.Generator.from_seed(2020)
x = g.normal(shape = (10,))
y = 3*x-2

print('x :',x.numpy())
print('y :',y.numpy())

#손실 함수 정의
def cal_mse(x,y,a,b):
    y_pred = a*x+b
    squared_error = (y_pred - y)**2
    mean_squared_error = tf.reduce_mean(squared_error)
    return mean_squared_error


a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHS = 200

for epoch in range(1, EPOCHS + 1):
    with tf.GradientTape() as tape:
        mse = cal_mse(x,y,a,b)
    grad = tape.gradient(mse,{'a':a,'b':b})
    d_a,d_b = grad['a'],grad['b']
    print('1    EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
    
    a.assign_sub(d_a*0.05)
    b.assign_sub(d_b*0.05)

    print('2    EPOCH %d - MSE : %.4f ----- a : %.2f ------ b : %.2f'%(epoch, mse, a, b))
```

```python
#경사하강법 실습

#샘플에 활용할 데이터 셋 만들기
def make_linear(w, b, size,noise):
    x = np.random.rand(size)
    y = w*x+b
    noise = np.random.uniform(-abs(noise),abs(noise),size = y.shape)
    yy = y + noise
    plt.figure(figsize=(10, 7))
    plt.plot(x,y,color='r',label=f'y={w}*x+{b}')
    plt.scatter(x,yy,label ='data')
    plt.legend(fontsize=20)
    plt.show()
    print(f'w:{w},b:{b}')
    return x,yy

x,y = make_linear(w=0.3,b=0.5,size=100,noise=0.01)

#최대 반복 횟수
num_epoch = 100

#학습률(learnig_rate)
learning_rate = 0.005

#에러기록
errors = []

#random 한 값으로 w,b를 초기화
w = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low = 0.0, high = 1.0)

for epoch in range(num_epoch):
    #Hypothesis 정의
    y_hat = w*x+b
    
    #Loss Function 정의
    error = 0.5*((y_hat - y)**2).sum()
    if error < 0.005:
        break
    #Gradient 미분 계산
    w = w - learning_rate*((y_hat - y)*x).sum()
    b = b - learning_rate*(y_hat -y).sum()
    
    errors.append(error)
    
    if epoch%5 == 0:
        print("{0:2}     w = {1:.5f}, b = {2:.5f}     error = {3:.5f}".format(epoch, w, b, error))

print("----" * 15)
print("{0:2}     w = {1:.1f}, b = {2:.1f}     error = {3:.5f}".format(epoch, w, b, error))

plt.figure(figsize = (10,7))
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

#케라스의 내장 데이터셋에서 mnist 다운
mnist = tf.keras.datasets.mnist

#load_data()로 데이터셋 로드
(x_train, y_train),(x_test,y_test) = mnist.load_data()

#로드 된 데이터셋 확인
print('train set', x_train.shape, y_train.shape)
print('test set', x_test.shape, y_test.shape)

#데이터 시각화
#canvas 생성
fig, axes = plt.subplots(3,5)
fig.set_size_inches(8,5)

for i in range(15):
    ax = axes[i//5,i%5]
    #imshow 로 이미지 시각화
    ax.imshow(x_train[i], cmap='gray')
    ax.axis('off')
    ax.set_title(str(y_train[i]))

plt.tight_layout()
plt.show()
```

```python
#데이터 전처리(정규화)
#데이터 전처리 전 데이터의 형태 확인
x_train[0, 10:15, 10:15,]

print(f'정규화 전] 최소값:{x_train.min()}, 최대값:{x_train.max()}')

#데이터 정규화(0~1사이의 값으로 정규화)
x_train = x_train/x_train.max()
print(f'정규화 후] 최소값:{x_train.min()}, 최대값:{x_train.max()}')

#테스트 데이터 셋에도 동일하게 정규화
x_test = x_test/x_test.max()

#변환 후 x_train 배열의 데이터 확인
x_train[0, 10:15, 10:15,]

#데이터 정규화를 하는 이유는 입력데이터가 정규화 되어 모델이 학습하는 경우에 경사하강법 알고리즘에 의한 수렴속도가 빠르고, 국소 최적(local optimum)에 빠지는 현상을 방지해주기 때문이다.
```

```python
#심층 신경망으로 이미지 분류 정리
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

#케라스의 내장 데이터셋에서 mnist 다운
mnist = tf.keras.datasets.mnist

#load_data()로 데이터셋 로드
(x_train, y_train),(x_test,y_test) = mnist.load_data()

#로드 된 데이터셋 확인
#print('train set', x_train.shape, y_train.shape)
#print('test set', x_test.shape, y_test.shape)

#데이터 정규화
#데이터 정규화를 하는 이유는 입력데이터가 정규화 되어 모델이 학습하는 경우에 경사하강법 알고리즘에 의한 수렴속도가 빠르고, 국소 최적(local optimum)에 빠지는 현상을 방지해주기 때문이다.
x_train = x_train/x_train.max()
x_test = x_test/x_test.max()

model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28,28)), #Dense 레이어에 입력값으로 넣기 위해서는 (28,28)의 2차원 데이터로 넣을수 없기 때문에 (784)의 1차원 데이터로 변경
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(32,activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax') #노드 = 10개(클래스 개수와 동일하게 : mnist 데이터는 0~9니깐)
]) 

model.compile(optimizer='adam', #옵티마이저 : 손실을 낮추기 위해서, 신경망의 가중치와 학습률과 같은 신경망 속성을 변경하는 최적화 방법
              loss='sparse_categorical_crossentropy',
              metrics =['accuracy'])

#훈련
model.fit(x_train, y_train,
          #검증 셋 지정
          validation_data=(x_test,y_test),
          epochs=10)

#평가
test_loss, test_acc = model.evaluate(x_test,y_test)


#예측
predictions = model.predict(x_test)

#이미지, 원래 값, 예측값, 신뢰도 출력
def get_one_result(idx):
    img, y_true, y_pred, confidence = x_test[idx], y_test[idx], np.argmax(predictions[idx]),100*np.max(predictions[idx])
    return img, y_true, y_pred, confidence

#canvas 생성
fig, axes = plt.subplots(3,5)
fig.set_size_inches(12,10)

for i in range(15):
    ax = axes[i//5,i%5]
    img, y_true, y_pred, confidence = get_one_result(i)
    #imshow 로 이미지 시각화
    ax.imshow(img, cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(f'True:{y_true}')
    ax.set_xlabel(f'Prediction:{y_pred}\nConfidence:({confidence:.2f} %)')

plt.tight_layout()
plt.show()
```
