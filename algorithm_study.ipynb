```python
#그리디_거스름돈
n = int(input())

money_list = [500,100,50,10]

coin_count = 0

for i in money_list:
    if n >= i:
        coin_count = coin_count + n//i
        n = n%i

print(coin_count)
        
```

```python
#그리디_큰 수의 법칙
#단순 반복 계산
'''
n, m, k = map(int, input().split())
num_list = list(map(int, input().split()))

num_list.sort(reverse=True)

result = 0
count = 0
while m != 0:
    if count == k:
        result = result + num_list[1]
        count = 0
        m = m-1
    else:
        result = result + num_list[0]
        count = count + 1
        m = m-1
        
print(result)
'''

#시간 단축
n, m, k = map(int, input().split())
num_list = list(map(int, input().split()))
num_list.sort(reverse=True)

multiplication = m//(k+1)
addition = m%(k+1)

result = multiplication*(num_list[0]*k+num_list[1]) + addition*num_list[0]

print(result)
```

```python
#그리디_숫자 카드 게임
#내 답
'''
n, m = map(int, input().split())
map_card = []
for i in range(n):
    map_card.append(list(map(int, input().split())))

pick_card = []
for i in range(n):
    pick_card.append(min(map_card[i]))

print(max(pick_card))
'''

#답지
n, m = map(int, input().split())
pick_card = 0
for i in range(n):
    pick_card = max(min(list(map(int, input().split()))),pick_card)
print(pick_card)
```

```python
#그리디_1이 될 때까지
n, k = map(int, input().split())

count = 0
while n >= k:
    if n%k == 0:
        n = n/k
        count += 1
    else :
        count += n%k
        n -= n%k
if n != 1:
    count += (n-1)

print(int(count))
```

```python
#구현_상하좌우

n = int(input())
plan = list(map(str, input().split()))

x, y = 1,1
d = ['L','R','U','D']
dx = [0,0,-1,1]
dy = [-1,1,0,0]

for i in plan:
    for j in range(4):
        if d[j] == i:
            if x+dx[j] > 0 and x+dx[j] < n + 1:
                x = x + dx[j]
            if y+dy[j] > 0 and y+dy[j] < n + 1:
                y = y + dy[j]

print(x,y)
```
