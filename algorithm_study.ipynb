```python
#그리디_1이 될 때까지
n, k = map(int, input().split())

count = 0
while n >= k:
    if n%k == 0:
        n = n/k
        count += 1
    else :
        count += n%k
        n -= n%k
if n != 1:
    count += (n-1)

print(int(count))
```

```python
#BFS
from collections import deque

graph =[[],
        [2,3,8],
        [1,7],
        [1,4,5],
        [3,5],
        [3,4],
        [7],
        [2,6,8],
        [1,7]]

visited = [0]*9

def bfs(x):
    queue = deque([x])
    visited[x] = 1
    print(x,end = ' ')
    while len(queue) != 0:
        y = queue.popleft()
        for i in graph[y]:
            if visited[i] == 0:
                queue.append(i)
                visited[i] = 1
                print(i, end = ' ')
                
bfs(1)
print('')
print(visited)
```
