```python
a, b = 6, 30
array = []

for i in range(a,b+1):
    if i**0.5 == int(i**0.5):
        array.append(int(i**0.5))
    elif i**(1/3) == int(i**(1/3)):
        array.append(int(i**(1/3)))

count = 0
for i in array:
    check = 0
    for j in range(1,i):
        if j == 1:
            pass
        else:
            if i%j == 0:
                check = 1
                break
    if check == 0:
        count += 1

print(count)



```
